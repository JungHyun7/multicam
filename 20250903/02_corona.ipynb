```python
# DataFrame 안에 cloumns 변수에 데이터를 대입
# 유의할 점: 기존의 컬럼의 개수와 같은 1차원 데이터를 대입
corona.columns = ['등록일시',
                '총사망자',
                '총확진자',
                '기준일',
                '기준시간',
                '수정일시',
                '누적의심자',
                '누적확진율'
]
```

```python
# 데이터프레임에서 정렬을 변경하는 방법
# 정렬 (sort) -> list에서는 사용이 가능
# 정렬을 변경할 수 있는 기준이 2개 -> index, values
# sort_index(), sort_values() 2개가 모두 존재
# sort_values({col_name}, ascending = {True | False}, inplcae = {False | True})
# 등록일시를 기준으로 현재는 내림차순 정렬을 오름차순 정렬로 변경

corona.sort_values(['등록일시'],
                ascending = True,
                inplace= True
)
```

```python
# 컬럼의 순서를 오름차순 변경?
# sort() 함수가 존재x
# columns 데이터를 리스트변경 -> sort()로 데이터를 변경 -> 필터를 이용해서 순서를 변경
col_list = list(
    corona.columns
)
col_list.sort()
corona[col_list]
```

```python
# 복사본 생성 : backup -> 1차 작업이 완료하고 중간 저장
df = corona.copy()
```

```python
# 결측치가 존재하는 컬럼이 3개 -> 수정일시, 누적의심자, 누적확진율
# 결측치가 존재하는 데이터를 확인 (수정일시)
# 수정일시 데이터 중 결측치가 존재 유무
# df 데이터프레임에서 특정 컬럼 하나를 선택 -> 2차원 데이터프레임에서 1차원 시리즈로 변경
# 인덱스의 조건식

flag = df['수정일시'].isna()
```

```python
# 인덱스의 조건을 이용하여 데이터프레임 필터
# DataFrame.loc[인덱스의 조건, 컬럼의 조건]
# df.loc[~flag]
# 비트 연산자 '~'를 통해서 1차원 논리값의 부정 표현
# '-' 연산자를 이용하여 1차원 논리값의 부정 표현
df.loc[-flag]
```

```python
# 누적확진율 데이터에서 결측치인 데이터만 확인
flag2 = df['누적확진율'].isna()
```

```python
# 인덱스를 초기화
# 초기화(reset) + 위치(index)
# reset_index() -> 인덱스를 재지정하고 기존의 인덱스를 컬럼의 데이터로 변환
# 매개변수 inplace -> 원본의 데이터를 수정할 것인가?
# 매개변수 drop -> 기존의 인덱스를 제거할 것인가?
df.reset_index(drop=True, inplace=True)
```

```python
# 총확진자수 데이터를 이용해서 일일확진자 데이터를 생성
# 기존의 데이터를 이용하여 새로운 데이터를 생성 -> 파생변수
# 기존의 데이터만으로 설명이 불가능하거나 아니면 상관 관계가 약한경우
# 새로운 데이터를 생성해서 설명이 가능하고 상관관계를 올리수 있게 하는 방법
# 일일확진자 -> 오늘의 총확진자수 - 어제의 총확진자수
df.loc[1, '총확진자'] - df.loc[0, '총확진자']
```

```python
df.loc[2, '총확진자'] - df.loc[1, '총확진자']
df.loc[3, '총확진자'] - df.loc[2, '총확진자']
df.loc[4, '총확진자'] - df.loc[3, '총확진자']
df.loc[5, '총확진자'] - df.loc[4, '총확진자']
#....
df.loc[819, '총확진자'] - df.loc[818, '총확진자']
```

```python
#  반복문을 이용하여 일일확진자수를 리스트로 생성
new_list = [0]

#반복문 시작 (데이터프레임의 길이에 -1 반복 실행)
for i in range(1, len(df)):
    # -> 1부터 819 대입하여 반복
    new_list.append(df.loc[i, '총확진자'] - df.loc[i-1, '총확진자'])
    
len(new_list)
```

```python
# df에 새로운 컬러 '알일확진자'를 추가하고 데이터는 new_list 대입
# dict에서 key : value를 추가하는 형태와 같다
df['일일확진자'] = new_list
df
```

```python
# index를 이동하는 함수
# shift(n) -> n만큼 인덱스를 변경 n의 기본값은 1
df['일일사망자'] = df['총사망자'] - df['총사망자'].shift()
```

```python
df
```

```python
# 특정 시리즈 안에서의 차이를 구하는 함수
# diff()  n : 인덱수 수 만큼
df['총확진자'].diff()
```

```python
# 일일사망자의 결측치를 0으로 채워준다.
df['일일사망자'].fillna(0, inplace=True)
```

```python
# 일일사망자, 일일확진자 데이터는 범위가 0보다 크거나 같은 데이터여야 한다.
# 비교 연산자를 이용하여 해당 데이터를 확인
flag = df['일일사망자'] < 0
```

```python
df.loc[flag]
```

```python
# 일일확진자가 음수인 데이터를 확인
flag - df['일일확진자'] < 0
df.loc[flag]
```

```python
# 일일확진자, 일일사망자 두개의 컬럼을 모두 선택해서 비교
# df [['일일확진자', '일일사망자']] < 0
# flag = df [['일일확진자', '일일사망자']] < 0
# 인덱스의 조건식에 1차원 데이터 형이 아닌 2차원 데이터가 들어가면 에러가 발생
# df.loc[flag]
```

```python
# '일일확진자', '일일사망자' 컬럼을 제거
df.drop(
    ['일일확진자', '일일사망자'],
    axis = 1,
    inplace = True
)
```

```python
# 일일확진자 컬럼은 shift() 함수를 이용하여 생성
df['일일확진자'] = df['총확진자'] - df['총확진자'].shift()
```

```python
# 일일사망자 컬럼은 diff() 함수를 이용하여 생성
df['일일사망자'] = df['총사망자'].diff()
```

```python
# 결측치 제거
df['일일확진자'].fillna(0, inplace=True)
df['일일사망자'].fillna(0, inplace=True)
```
