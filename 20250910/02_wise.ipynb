```python
import pandas as pd
from bs4 import BeautifulSoup as bs
from selenium import webdriver
```

```python
# 접속 주소
url = "https://comp.wisereport.co.kr/company/c1010001.aspx?cmp_cd="
# 종목의 코드 (하이닉스 종목코드)
input_code = "000660"
```

### 연습
- webdriver의 Chorme을 이용하여 url + input_code로 요청
- driver의 html 문서를 불러온다.
- 불러온 html을 BeautifulSoup을 이용하여 데이터 파싱
- div 태그 중 class 이름이 "cmp_comment"와 같은 태그를 선택
    - li 태그들을 모두 찾아서 텍스트 추출하여 새로운 리스트를 생성(cmp_info)
- div 태그 중 class의 이름이 "cmp_comment_etc"와 같은 태그를 선택
    - li 태그들을 모두 찾아서 텍스트 추출하여 새로운 리스트를 생성(cmp_etc)
- DataFrame 생성해서 cmp_info, cmp_etc를 대입


## 수업 ver

```python
driver = webdriver.Chrome()
```

```python
# url + input_code를 이용해서 요청
driver.get(url+input_code)
```

```python
soup = bs(
    driver. page_source, 'html.parser'
)
```

```python
driver.quit()
```

```python
# div 태그 중 class가 'cmp_comment'인 태그를 선택
div_data = soup.find(
    'div',
    attrs = {
        'class' : 'cmp_comment'
    }
)
div_data
```

```python
# div_data에서 li 태그들을 모두 찾는다.
li_list = div_data.find_all('li')
```

```python
cmp_info = []
for li_data in li_list:
    # li_data는 Tag data 대입
    cmp_info.append(li_data.get_text())
cmp_info
```

```python
cmp_etc = list(
    map(
        lambda x : x.get_text(),
        soup.find(
            'div',
            attrs = {
                'class' : 'cmp_comment_etc'
            }
        ).find_all('li')
    )
)
```

```python
cmp_etc
```

```python
df1 = pd.DataFrame(
    [
        cmp_info, cmp_etc
    ], index = ['cmp_info', 'cmp_etc']
)
```

```python
df1
```

```python
df2 = pd.DataFrame(
    {
        'cmp_info' : cmp_info,
        'cmp_etc' : cmp_etc
    }
)
```

```python
df2['code'] = input_code
```

```python
df2
```

```python
# df1의 인덱스를 컬럼으로 컬럼을 인덱스로 변환하는 장치 -> 전치
df1.T
```

```python
# 데이터프레임의 행 결합의 조건 -> 데이터프레임의 구조가 같다
# pandas 안에 존재하는 concat() 함수를 이용
# concat() -> 단순한 데이터프레임의 결합
    # 첫번째 인자 -> 데이터프레임들 [df1, df2, df3, ...]
    # axis 매개변수 -> 행으로 결합할 것인가, 열로 결합 할것인가?
    # ingore_index -> 결합 후 인덱스를 초기화할 것인가?
```

```python
res = requests.get(url+input_code)
```

```python
soup = bs(res.text, 'html.parser')
```

```python
cmp_info = list(
    map(
        lambda x : x.get_text(),
        soup.find(
            'div', attrs = {'class' : 'cmp_comment'}
        ).find_all('li')
    )
)
cmp_etc = list(
    map(
        lambda x : x.get_text(),
        soup.find('div', attrs = {
            'class' : 'cmp_comment_etc'
        }).find_all('li')
    )
)
```

```python
df3 = pd.DataFrame({'cmp_info' : cmp_info, 'cmp_etc': cmp_etc})
df3['code'] = input_code
```

```python
df3
```

```python
pd.concat(
    [df2, df3], axis = 0, ignore_index = True
)
```
